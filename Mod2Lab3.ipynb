```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

# Step 1: Create lists for seasons and stats (PPG)
seasons = [
    "1996-97", "1997-98", "1998-99", "1999-00", "2000-01", "2001-02", "2002-03",
    "2003-04", "2004-05", "2005-06", "2006-07", "2007-08", "2008-09", "2009-10",
    "2010-11", "2011-12", "2012-13", "2013-14", "2014-15", "2015-16"
]

ppg = [
    7.6, 15.4, 19.9, 22.5, 28.5, 25.2, 30.0, 24.0, 27.6, 35.4,
    31.6, 28.3, 26.8, 27.0, 25.3, 27.9, 27.3, 13.8, 22.3, 17.6
]

# Function to calculate top season and stat
def find_top_season_and_stat(seasons, stats):
    """
    Function to identify the top season and its corresponding stat.
    :param seasons: List of seasons.
    :param stats: List of stats corresponding to each season.
    :return: Tuple containing (top season, top stat).
    """
    # Sort the stats in descending order
    sorted_stats = sorted(stats, reverse=True)
    print("Sorted Stats (Highest to Lowest):")
    print(sorted_stats)
    
    # Find the top stat
    top_stat = sorted_stats[0]
    
    # Find the season corresponding to the top stat
    for i in range(len(stats)):
        if stats[i] == top_stat:
            top_season = seasons[i]
            break
    
    # Return the top season and stat as a tuple
    return top_season, top_stat

# Call the function and unpack the results
top_season, top_stat = find_top_season_and_stat(seasons, ppg)

# Print the results
print(f"\nKobe Bryant's top scoring season was {top_season} with {top_stat} PPG.\n")

# Produce a list of seasons ordered by sorted PPG
seasons_sorted_by_ppg = [season for _, season in sorted(zip(ppg, seasons), reverse=True)]

# Print the results
print("Seasons in order of PPG (highest to lowest):")
for season, ppg_stat in zip(seasons_sorted_by_ppg, sorted(ppg, reverse=True)):
    print(f"Season {season}: {ppg_stat} PPG")
```
