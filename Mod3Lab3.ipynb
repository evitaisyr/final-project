```python
import numpy as np

# List of MVP winners from 1999 to 2023
mvp_players = np.array([
    "Kurt Warner", "Marshall Faulk", "Kurt Warner", "Rich Gannon", "Steve McNair", "Peyton Manning",
    "Peyton Manning", "Shaun Alexander", "LaDainian Tomlinson", "Tom Brady", "Peyton Manning",
    "Peyton Manning", "Aaron Rodgers", "Adrian Peterson", "Peyton Manning", "Aaron Rodgers",
    "Cam Newton", "Matt Ryan", "Tom Brady", "Patrick Mahomes", "Lamar Jackson", "Aaron Rodgers",
    "Aaron Rodgers", "Patrick Mahomes", "Lamar Jackson", "Tom Brady"
])

# Find unique players
unique_players = np.unique(mvp_players)

# Approach 1: Count MVPs by looping through unique player names
print("MVP Awards Count (Approach 1):")
print("Player                 | MVP Awards")
print("-----------------------|------------")
for player in unique_players:
    count = np.count_nonzero(mvp_players == player)
    print(f"{player:<23} | {count:>10}")

print("\nMVP Awards Count (Approach 2):")
print("Player                 | MVP Awards")
print("-----------------------|------------")

# Approach 2: Count MVPs using indices of unique players
for i in range(len(unique_players)):
    count = np.count_nonzero(mvp_players == unique_players[i])
    print(f"{unique_players[i]:<23} | {count:>10}")

print("\nMVP Awards Sorted by Number of MVPs:")
print("Player                 | MVP Awards")
print("-----------------------|------------")

# Calculate MVP counts and sort by number of awards (descending)
mvp_counts = np.array([np.count_nonzero(mvp_players == player) for player in unique_players])
sorted_indices = np.argsort(-mvp_counts)  # Sort in descending order

# Print sorted results
for i in sorted_indices:
    print(f"{unique_players[i]:<23} | {mvp_counts[i]:>10}")

# Add a note about co-MVPs
print("\nNote: There were 25 seasons analyzed, but 26 MVP awards due to co-MVPs in 2003.")
```
